```python
# faccio un check sulle righe per dare un' occhiata ai valori nulli con il ciclo for  
for column in df.columns:
    nan_count = df[column].isna().sum()
    nan_percentage = round((nan_count/df.shape[0])*100,2)
    print(f"{nan_count}Nan values, {nan_percentage}% of all rows in column '{column}")
```

```python
# In questo passaggio ho raggruppato per continenti con la funzione groupby e fatto la somma totale per continenti di tutti i casi totali di contagiati,
# escludendo dal raggruppamento tutti quei paesi che non hanno una locazione definita.
df = df[df["continent"] != "N/D"]
def new_func(df):
    cases_for_continent = df.groupby("continent")["total_cases"].sum()
    return cases_for_continent

cases_for_continent = new_func(df)
print(cases_for_continent)
```

```python
# Prima mi sono calcolato la somma totale dei contagiati nel modo e dopo ho calcolato la percentuale di ogni singolo continente 
total_world_cases = df['total_cases'].sum()
cases_for_continent = df.groupby("continent")["total_cases"].sum()
percentage_for_continent = (cases_for_continent / total_world_cases) * 100
print(percentage_for_continent)
```

```python
# In questo passaggio ho calcolato i casi totali massimi,la media e la percentuale stampandoli a schermo 
total_world_cases = df['total_cases'].sum()
results = {}

for continent in df['continent'].unique():
    continent_df = df[df['continent'] == continent]
    max_cases = continent_df['total_cases'].max()
    mean_cases = continent_df['total_cases'].mean()
    percentage_for_continent = (continent_df['total_cases'].sum() / total_world_cases) * 100
    results[continent] = {'max_cases': max_cases, 'mean_cases': mean_cases, 'percentage_for_continent': percentage_for_continent}

for continent, data in results.items():
    print(f"Continent: {continent}")
    print(f"Max cases: {data['max_cases']}")
    print(f"Mean cases: {data['mean_cases']}")
    print(f"Percentage of cases in the world: {data['percentage_for_continent']}%\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df["date"] = pd.to_datetime(df["date"])
italy_2022 = df[(df["location"]== "Italy")&(df["date"].dt.year==2022)].copy()
plt.plot(italy_2022['date'], italy_2022['total_cases'], label='Casi Totali')
plt.xlabel("date")
plt.ylabel("total_cases")
plt.title("Andamento Covid-19 nel 2022 in Italia")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.plot(italy_2022['date'], italy_2022['new_cases'], label='Nuovi Casi')
plt.xlabel("date")
plt.ylabel("new cases")
plt.title("Andamento Covid-19 nel 2022 in Italia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
italy_2022['new_cases_cumsum'] = italy_2022['new_cases'].cumsum()
plt.plot(italy_2022['date'], italy_2022['new_cases_cumsum'], label='Somma cumulativa')
plt.xlabel("date")
plt.ylabel("new cases_cumsum")
plt.title("Somma cumulativa nuovi casi Covid-19 nel 2022 in Italia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Applico filto per le quattro nazioni dell'anno 2023 
year_2023 = df[df["location"].isin(["Italy","Germany","France","Spain"]) & (df["date"].dt.year == 2023)]      #hosp_patients
```

```python
# Dopo raggruppo per nazioni e sommo la colonna pazienti ospedalizzati e succesivamente con un ciclo for per ciclare non solo su hosp,
# ma anche sugli indici di country ed i valori count di hosp

hosp = year_2023.groupby("location")["hosp_patients"].sum()
for country, count in hosp.items():
    if pd.isnull(count):
        print(f"Non è possibile calcolare il numero totale dei pazienti ospedalizzati per {country} perchè ci sono dati nulli.")
    else:
        print(f"Il totale dei pazienti ospedalizzati per {country} nel 2023 è: {count}")
```
